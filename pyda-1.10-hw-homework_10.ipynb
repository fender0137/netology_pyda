*Задание 1*

Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
rating = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')


print(rating.head(10))
movies.head(10)
```

```python
data = pd.merge(rating, movies, left_on='movieId', right_on='movieId', how='left').sort_values(by='rating',ascending=False)
```

```python
raiting_films = data.filter(items = ['rating', 'title'])
films_5 = raiting_films[(raiting_films['rating']==5)]
```

```python
count_5_stars_film = films_5.groupby('title').count()
count_5_stars_film.sort_values('rating',ascending=False).head(1)
```

*Задание 2*

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
power.head(10)
country_films = power[(power['year']>=2005) & (power['year']<=2010) & 
                      ((power['category']==4) | (power['category']==12) | (power['category']==21))
                     & ((power['country']=='Lithuania')|(power['country']=='Latvia')|(power['country']=='Estonia'))]
country_films
```

```python
country_films.filter(items = ['quantity']).sum()
```
